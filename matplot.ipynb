```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
'Manufacturer': ['Toyota', 'Ford', 'Ford', 'Audi', 'BMW', 'Toyota', 'Audi', 'BMW',
'Nissan', 'Nissan'],
'Type': ['Sedan', 'Truck', 'SUV', 'Sedan', 'Sedan', 'Truck', 'SUV', 'Truck',
'SUV', 'Sedan'],
'Fuel_Efficiency': [34, 22, 24, 27, 28, 24, 26, 21, 23, 35],
'Horsepower': [180, 320, 260, 220, 240, 290, 210, 310, 250, 200],
'Price': [30000, 40000, 35000, 45000, 50000, 32000, 47000, 48000, 33000,
31000]
}
```

```python
df = pd.DataFrame(data)
df.sort_values('Horsepower',inplace=True)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(10,10))
axes[0][0].plot(df['Horsepower'],data['Price'],linestyle='--',color='red',label='Horsepower vs Price',marker='o')
axes[0][0].set_xlabel('Horsepower')
axes[0][0].set_ylabel('Price')
axes[0][0].legend(loc='upper right')
axes[0][0].set_xlim(df['Horsepower'].min()-20,df['Horsepower'].max()+20)
axes[0][0].set_ylim(df['Price'].min()-10000,df['Price'].max()+10000)
avg_per_type = df.groupby('Type')['Price'].mean().reset_index()
"""axes[1].plot(avg_per_type['Type'],avg_per_type['Price'],marker='o')
axes[1].set_xlabel('car type')
axes[1].set_ylabel('avrage price $')
axes[1].set_ylim(avg_per_type['Price'].min()-1000,avg_per_type['Price'].max()+1000)
"""
avg_horsepower_per_type = df.groupby('Type')['Horsepower'].mean().reset_index()
print(avg_horsepower_per_type)
axes[1][0].bar(avg_horsepower_per_type['Type'],avg_horsepower_per_type['Horsepower'],color='#E55DCA')
axes[0][1].hist(df['Fuel_Efficiency'],5,color='#E9E5BC')
axes[1][1].scatter(df['Horsepower'],df['Price'])
plt.title('try')
plt.tight_layout()
plt.savefig('my_fig.jpg')
plt.show()
```
