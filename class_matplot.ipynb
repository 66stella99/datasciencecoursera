```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import files
data_ = files.upload() # Select your kaggle.json file
df = pd.read_csv('matplotlib_subplot_class_exc (3).csv')
```

```python
df.to_string()
```

```python

df.sort_values(['Country','Year'],inplace=True)
df_fr_gdp=df[df['Country']=='France']
```

```python
df_fr_gdp
```

```python
fig, axes = plt.subplots(nrows=4,ncols=1,figsize=(10,20))
axes[0].plot(df_fr_gdp['Year'],df_fr_gdp['GDP per Capita'],label = 'GDP France',marker='o')
year_gdp = df[df['Year']==2005]
axes[1].hist(year_gdp['GDP per Capita'],5,rwidth = 0.2, label = 'GDP 2005',color = '#50ECF7')
gdp_2010 = df[df['Year']==2010]
axes[2].bar(gdp_2010['Country'],gdp_2010['GDP per Capita'],width=0.2,label = 'GDP 2010',color='#F7F250')
uk_gdp =df[df['Country']=='UK']
axes[3].scatter(uk_gdp['Year'],uk_gdp['GDP per Capita'],label = 'UK',color = '#F7509E')
for i,xy_label in enumerate([['Year','GDP'],['GDP','#'],['Country','GDP'],['Year','GDP']]):
    axes[i].legend()
    axes[i].set_xlabel(xy_label[0])
    axes[i].set_ylabel(xy_label[1])
plt.tight_layout()
plt.show()
```
