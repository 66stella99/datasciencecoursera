```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'a':[None,2,3,None,5],
                 'b':[10,None,None,40,50],
                 'c':[100,None,300,400,None]})
```

```python
df.dropna(axis=0)
df.dropna(how='all')
df.dropna(subset=['a','c'])
df.replace(np.nan,df.mean())
df['b'].replace(np.nan,df['b'].mean())
```
